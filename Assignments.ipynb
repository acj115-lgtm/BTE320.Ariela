```python
import random

play_again = input("Do you want to play Rock-Paper-Scissors? (yes/no): ").lower()

while play_again == "yes":
    user_choice = input("Enter rock, paper, or scissors: ").lower()
    computer_choice = random.choice(['rock', 'paper', 'scissors'])

    print(f"Computer chose: {computer_choice}")

    if user_choice == computer_choice:
        print("It's a tie!")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win!")
    else:
        print("Computer wins!")

    play_again = input("Play again? (yes/no): ").lower()

print("Thanks for playing!")
```

```python
class Student:
  def __init__(self, name, number):
    self.name = name
    self.number = number
    self._courses = []

  def get_name(self):
    return self.name

  def get_courses(self):
    return self._courses

  def set_name(self, newName):
    assert isinstance(newName, str)
    self._name = newName

  def __len__(self):
    return len(self._courses)

  def enroll(self, newCourse):
    if newCourse not in self._courses:
      self._courses.append(newCourse)
    else:
      print(f"Student {self.name} is already enrolled in {newCourse}")
```

```python
class Point:
  def __init__(self, x, y):
      self.x = x
      self.y = y

  def distance(self, other):
      dx = (self.x - other.x) ** 2
      dy = (self.y - other.y) ** 2
      return (dx + dy) ** 0.5

class Rectangle:
    def __init__(self):
        self.p1 = Point(0,0)
        self.p2 = Point(5,0)
        self.p3 = Point(0,5)
        self.p4 = Point(5,5)
        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height

```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.hourly_wage

  def set_wage(self, newWage):
    if newWage <= 0:
      print("Invalid wage value.")
    else:
      self.hourly_wage = new_wage


  def calculate_pay(self):
      return self.hourly_wage * self.hours

  def display_info(self):
    print(f"Employee: {self.name}\nHourly Wage: {self.hourly_wage}")

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly payment is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly payment is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus")
```
