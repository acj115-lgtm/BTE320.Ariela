```python
import math

ticket_price = 10
fixed_costs = 200

print(f"{'Ad$':<8} {'Profit':<8}")
for ad in range(0, 201, 25):
    # Additional attendees = 2 * advertisingDollars^0.5
    attendees = 20 + (2 * math.sqrt(ad))
    revenue = attendees * ticket_price
    total_costs = fixed_costs + ad
    profit = revenue - total_costs
    print(f"{float(ad):<8} {float(profit):<8}")
```

```python
def loan_payment(principal, annual_rate, years):
    # Validate positive inputs
    if principal <= 0 or annual_rate <= 0 or years <= 0:
        return "Error: All inputs must be positive numbers."

    r = (annual_rate / 100) / 12  # monthly interest rate
    n = years * 12                # total monthly payments

    payment = (r * principal) / (1 - (1 + r) ** -n)
    return payment

# Example call
print(f"Monthly Payment: ${loan_payment(10000, 5, 3):.2f}")
```

```python
import random

play_again = input("Do you want to play Rock-Paper-Scissors? (yes/no): ").lower()

while play_again == "yes":
    user_choice = input("Enter rock, paper, or scissors: ").lower()
    computer_choice = random.choice(['rock', 'paper', 'scissors'])

    print(f"Computer chose: {computer_choice}")

    if user_choice == computer_choice:
        print("It's a tie!")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win!")
    else:
        print("Computer wins!")

    play_again = input("Play again? (yes/no): ").lower()

print("Thanks for playing!")
```

```python
def roi(cost, revenue):
  assert cost != 0
  assert isinstance(cost, (int, float))
  assert isinstance(revenue, (int, float))
  return (revenue - cost) / cost

cost = float(input("Enter cost: "))
revenue = float(input("Enter revenue: "))
print(roi(cost, revenue))
```

```python
def loan_payment(principal, annual_rate, months):
    if principal <= 0 or annual_rate <= 0 or months <= 0:
        return "Error: Inputs must be positive numbers."

    r = (annual_rate / 100) / 12

    if r != 0:
        payment = (r * principal) / (1 - (1 + r) ** (-months))
    else:
        payment = principal / months

    return payment

result = loan_payment(10000, 5, 36)
print(f"Monthly Payment: ${result:.2f}")
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5

point1 = Point(0, 0)
point2 = Point(3, 4)
print(point1.distance(point2))
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5

class Rectangle:
    def __init__(self):
        self.p1 = Point(0, 0)
        self.p2 = Point(5, 0)
        self.p3 = Point(0, 5)
        self.p4 = Point(5, 5)
        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height

my_rect = Rectangle()
print(f"Width: {my_rect.width}")
print(f"Height: {my_rect.height}")
print(f"Area: {my_rect.area()}")
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.hourly_wage

  def set_wage(self, newWage):
    if newWage <= 0:
      print("Invalid wage value.")
    else:
      self.hourly_wage = new_wage


  def calculate_pay(self):
      return self.hourly_wage * self.hours

  def display_info(self):
    print(f"Employee: {self.name}\nHourly Wage: {self.hourly_wage}")

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly payment is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly payment is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus")
```
